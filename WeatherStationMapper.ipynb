```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
merged_df = pd.read_csv("climate_merged.csv")
# Remove columns we don't need
merged_df = merged_df[["LATITUDE", "LONGITUDE"]]
# Remove duplicates. This will make processing way faster
merged_df = merged_df.drop_duplicates()
```

```python
# Make the world map of the stations
gdf = gpd.GeoDataFrame(
    merged_df, geometry=gpd.points_from_xy(merged_df['LONGITUDE'], merged_df['LATITUDE']))

world_map = gpd.read_file("/Users/noah/VENVS/geoplotter/lib/python3.11/site-packages/pyogrio/tests/fixtures/naturalearth_lowres/naturalearth_lowres.shp")
# Replace the file path with the path to naturalearth_lowres.shp on your computer
# This requires geopandas to be installed
# It is recommended to do it in a virtial environment

world_map.plot(figsize=(10, 6))
gdf.plot(ax=world_map.plot(figsize=(10, 6)), color='red', markersize=1)
plt.show()

# Note: This code was heavily generated using Google Gemini.
```
