```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Wages By State.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
# Convert "Employment" to numeric (if not already)
df['Employment'] = pd.to_numeric(df['Employment'], errors='coerce')

# Calculate the mean of employment
mean_employment = df['Employment'].mean()

# Print the result
print(f"The mean of employment is: {mean_employment}")
```

```python
median_employment = df['Employment'].median()

# Print the result
print(f"The median of employment is: {median_employment}")
```

```python
# Create the bar chart for employment per state
plt.figure(figsize=(12, 6))
plt.bar(df['State'], df['Employment'], color='blue', alpha=0.7)

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Force the y-axis to display from 0 to 8000
plt.ylim(0, 8000)  # Explicitly set the y-axis range

# Add labels and title
plt.xlabel('States')
plt.ylabel('Employment')
plt.title('Bar Chart of Employment per State')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
# Plot the scatter plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.scatter(df['State'], df['Employment'], color='blue', alpha=0.7)

# Add title and labels
plt.title('Employment by State', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Employment (in thousands)', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Sort the data by State if necessary (optional)
df_sorted = df.sort_values('State')

# Plot the line graph
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(df_sorted['State'], df_sorted['Employment'], marker='o', color='green', linestyle='-', alpha=0.7)

# Add title and labels
plt.title('Employment by State (Line Graph)', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Employment (in thousands)', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```
